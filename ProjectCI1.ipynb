# Compound Interest Calculatot 

### Determine how much your money can grow using the power of compound interest.

```python
# ask user to enter principal amount $
# ask user to enter interest rate P.C% 
# ask user to enter number of periods 

while True:
    amount=int(input("Enter principal amount $  "))
    rate=int(input("enter interest rate P.C% "))
    if rate <= 100:
        period=int(input("enter Number of periods "))
        break
    else:
        print("it can not be over 100% ")


    
```

```python
# function that calculate CI

def compound_interest(amount, rate, period):
    Amount = amount
    for i in range(period):
        Amount = Amount * (1 + rate/100)
    CI = Amount - amount
    return CI

compound_interest(amount,rate , period)



# view rate or profit of each period in 

def ci(amount, rate, period):
        Amount = amount
        dic={}
        for i in range(period):
            Amount = Amount * (1 + rate/100)
            CI= Amount - amount
            dic={int(Amount) : int(CI)}
            print(i , dic)
            
        last_key = list(dic) [-1]
        print ("Invesement value =" + str(last_key),'$')
        print("Compound interest =",int(compound_interest(amount,rate , period)), '$')
        per=  ( (compound_interest(amount,rate , period)/ amount ) * 100 )
        print('Percentage profit = ',int(per),"%")
        
# total interesr/ earnings 
# percentage profit 
# invesment value        
#simple interest         
        

ci(amount,rate , period)  

# # Simple Interest Calculatot 

si = lambda amount ,period ,rate : (amount * period * rate)/100 

print("simple interest = ",si(amount,rate,period), "$")
```
